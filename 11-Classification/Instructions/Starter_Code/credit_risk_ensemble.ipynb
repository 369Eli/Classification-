## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

## Split the Data into Training and Testing

```python
# Create our features by spliting data into 80/20 with the Numpy Library
# change the objects to integers ex. home_ownership using encoder  
## Now lets change all the Catagorical Values (any text or objects) into numerical values  to see 
## by using obj_df.dtypes  ----   what types of data we have?
```

```python
df.info()
```

```python
# Drop the Loan Status Column bc it is our desired output. So we dont want to change that column.
```

```python
# Separate the Features (X) from the Target (y)
y = df["loan_status"] # Traget
X = df.drop(columns="loan_status") # Independent Variables - Predictor
X = pd.get_dummies(X,drop_first=True)
```

```python
X
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Check if all data is numerical and list the desired output of the y values or the loan_status
y.value_counts()
```

```python
# Split data into train and test subsets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
```

```python
# Create and train LR model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',  max_iter=200, random_state=1)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# FIT THE DATA
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='warn', n_jobs=None, penalty='l2',
                   random_state=1, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
```

```python
# Execute LR predictions
predictions = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head()
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
# Check the balance of our target values
confusion_matrix(y_test,predictions)
```

```python
print(classification_report_imbalanced(y_test,predictions))
```

```python
target_names = ["high_risk", "low_risk"]
print(classification_report(y_test, predictions, target_names=target_names))
balanced_accuracy_score(y_test,predictions)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Easy Ensemble Classifier

```python
# Train the Classifier
from imblearn.ensemble import EasyEnsembleClassifier
eef = EasyEnsembleClassifier(n_estimators=100, random_state=1)
eef.fit(X_train_scaled, y_train)
```
